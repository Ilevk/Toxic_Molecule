```python
import os
from os.path import join

import pandas as pd
import numpy as np
```

```python
train_file_path = join('data', 'train.csv')
```

```python
df = pd.read_csv(train_file_path)
```

```python
from sklearn.model_selection import train_test_split

train, valid = train_test_split(df,
                                test_size=0.2,
                                shuffle=True,
                                random_state=2020)
```
